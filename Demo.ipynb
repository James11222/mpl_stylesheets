```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Theory Curve
x_theory = np.linspace(0, 2 * np.pi, 100)
y_theory = np.sin(x_theory)

# Data Points
np.random.seed(10)
x_data = x_theory[::10]
y_data = np.sin(x_data) + (np.random.rand(len(x_data)) - 0.5)
_y = np.random.rand(len(x_data)) / 2


# Plotting with Default
plt.style.use("default")
f, ax = plt.subplots(1,1)
ax.plot(x_theory,y_theory)
ax.errorbar(x_data, y_data, yerr = _y, fmt="o")
ax.set_ylabel("$f(x)$")
ax.set_xlabel("$x$")
plt.savefig("./images/default.svg")
plt.show()

# Plotting with Default
plt.style.use("dark_background")
f, ax = plt.subplots(1,1)
ax.plot(x_theory,y_theory)
ax.errorbar(x_data, y_data, yerr = _y, fmt="o")
ax.set_ylabel("$f(x)$")
ax.set_xlabel("$x$")
plt.savefig("./images/dark_background.svg")
plt.show()

# Plotting with Light Mode
plt.style.use("light_mode.mplstyle")
f, ax = plt.subplots(1,1)
ax.plot(x_theory,y_theory)
ax.errorbar(x_data, y_data, yerr = _y, fmt="o")
ax.set_ylabel("$f(x)$")
ax.set_xlabel("$x$")
plt.savefig("./images/light_mode.svg")
plt.show()

# Plotting with Dark Mode
plt.style.use("dark_mode.mplstyle")
f, ax = plt.subplots(1,1)
ax.plot(x_theory,y_theory)
ax.errorbar(x_data, y_data, yerr = _y, fmt="o")
ax.set_ylabel("$f(x)$")
ax.set_xlabel("$x$")
plt.savefig("./images/dark_mode.svg")
plt.show()
```

```python
# Plotting with Light Mode
plt.style.use("light_mode.mplstyle")
f, ax = plt.subplots(1,1)
ax.plot(x_theory,y_theory)
ax.plot(x_theory,y_theory + 0.1)
ax.plot(x_theory,y_theory + 0.2)
ax.plot(x_theory,y_theory + 0.3)
ax.plot(x_theory,y_theory + 0.4)
ax.plot(x_theory,y_theory + 0.5)
ax.set_ylabel("$f(x)$")
ax.set_xlabel("$x$")
plt.show()

# Plotting with Dark Mode
plt.style.use("dark_mode.mplstyle")
f, ax = plt.subplots(1,1)
ax.plot(x_theory,y_theory)
ax.plot(x_theory,y_theory + 0.1)
ax.plot(x_theory,y_theory + 0.2)
ax.plot(x_theory,y_theory + 0.3)
ax.plot(x_theory,y_theory + 0.4)
ax.plot(x_theory,y_theory + 0.5)
ax.set_ylabel("$f(x)$")
ax.set_xlabel("$x$")
plt.show()
```
